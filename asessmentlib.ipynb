```python
import matplotlib.pyplot as plt

# Given data
x = [3, 4, 5, 6]
y = [1.5, 2, 2.5, 3]

# Create line plot
plt.plot(x, y, marker='*')

# Add labels
plt.xlabel("X")
plt.ylabel("Y")

# Add title
plt.title("Line Plot of X and Y")

# Display the graph
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create array from 0 to 6 with increment of 0.3
x = np.arange(0, 6.1, 0.3)

# Plot x, x², x³, and x⁴ with specified styles
plt.plot(x, x, "ro", label="x")
plt.plot(x, x**2, "bs", label="x²")
plt.plot(x, x**3, "g", label="x³")
plt.plot(x, x**4, ":", label="x⁴")

plt.xlim(0, 6)
plt.ylim(0, 125)

# Add labels and title
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Plots of x, x², x³, and x⁴")

# Show legend
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Given data
height = [179, 155, 191, 152, 188, 177]
names = ['QA', 'WB', 'EC', 'RD', 'TE', 'YF']

# Create bar plot
plt.bar(names, height)

# Add title and labels
plt.title("Height Comparison")
plt.xlabel("Names")
plt.ylabel("Height (cm)")

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100,000 random numbers from a normal distribution
x = np.random.randn(100000)

# Histogram with 10 bins
plt.hist(x, bins=10)
plt.title("Histogram with 10 Bins")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Histogram with 20 bins
plt.hist(x, bins=20)
plt.title("Histogram with 20 Bins")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Histogram with 50 bins
plt.hist(x, bins=50)
plt.title("Histogram with 50 Bins")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
15. Supervised and Unsupervised Machine Learning

Supervised Learning:
It is a type of machine learning that uses labeled data (input and correct output) to train
the model. It is mainly used for prediction and classification.

Unsupervised Learning:
It is a type of machine learning that uses unlabeled data. The model finds hidden patterns,
relationships, or groups in the data without predefined outputs.
```
